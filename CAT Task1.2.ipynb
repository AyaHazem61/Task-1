```python
s = input()
```

```python
def swap(s) :
    n = len(s)
    s1 = ""
    for i in range(0,n) :
        if s[i].isupper() :
            s1 += s[i].lower()
        elif s[i].islower :
            s1 += s[i].upper()
        else :
            s1 += s[i]
    return s1
```

```python
print(swap(s))
```
